## 几个基本的、常用的接口

- `D.calendar()`: 获取交易日历，默认是从数据库的最早时间到最晚时间，如果future=True，就一直到当前年年底。start_time=None, end_time=None, freq="day", future=False
- `D.instruments()`： 获取一个stock pool config，默认market='all'。market="all", filter_pipe=None
- `D.list_instruments(instruments=xxx)`: 列出符合某个stock pool config的股票列表（或者字典），如果as_list=True就是返回列表，否则就是字典，字典的值条目应该是那个股票上市的时间段。instruments, start_time=None, end_time=None, freq="day", as_list=False

# 获取某市场中的满足一些要求的股票

使用 NameDFilter 和 ExpressionFilter

```python
from qlib.data import D
from qlib.data.filter import NameDFilter
```

```python
nameDFilter = NameDFilter(name_rule_re='SH[0-9]{4}55') # 使用NameDFilter
```

```python
D.instruments(market='csi300', filter_pipe=[nameDFilter])
```

```python
instruments = D.instruments(market='csi300', filter_pipe=[nameDFilter])
D.list_instruments(instruments=instruments, start_time='2015-01-01', end_time='2016-02-15', as_list=True)
```

```python
from qlib.data import D
from qlib.data.filter import ExpressionDFilter
expressionDFilter = ExpressionDFilter(rule_expression='$close>2000')
instruments = D.instruments(market='csi300', filter_pipe=[expressionDFilter])
resList = D.list_instruments(instruments=instruments, as_list=True)
# TODO - 研究一下这里，为什么重复调用也时间很长？不是有Datacache吗？我的redis服务器应该已经起作用了呀
# 我知道了，D.list_instruments不会使用到ExpressionCache和DataCache，因为按理说这个不会反复调用；但是D.features就会使用到ExpressionCache和DataCache
```

表达式直接在fields列表里面输入就行

```python
from qlib.data import D
instruments = ['SH600000']
fields = ['$close', '$volume', 'Ref($close, 1)', 'Mean($close, 3)', '$high-$low']
D.features(instruments, fields, start_time='2010-01-01', end_time='2017-12-31', freq='day').head()
```

```python
from qlib.data import D
from qlib.data.filter import NameDFilter, ExpressionDFilter
nameDFilter = NameDFilter(name_rule_re='SH[0-9]{4}55')
expressionDFilter = ExpressionDFilter(rule_expression='$close>Ref($close,1)')
instruments = D.instruments(market='csi300', filter_pipe=[nameDFilter, expressionDFilter])
fields = ['$close', '$volume', 'Ref($close, 1)', 'Mean($close, 3)', '$high-$low']
D.features(instruments, fields, start_time='2010-01-01', end_time='2017-12-31', freq='day').head()
```

```python
# 重复调用很快
from qlib.data import D
from qlib.data.filter import NameDFilter, ExpressionDFilter
nameDFilter = NameDFilter(name_rule_re='SH[0-9]{4}55')
expressionDFilter = ExpressionDFilter(rule_expression='$close>Ref($close,1)')
instruments = D.instruments(market='csi300', filter_pipe=[nameDFilter, expressionDFilter])
fields = ['$close', '$volume', 'Ref($close, 1)', 'Mean($close, 3)', '$high-$low']
D.features(instruments, fields, start_time='2010-01-01', end_time='2017-12-31', freq='day').head()
```
